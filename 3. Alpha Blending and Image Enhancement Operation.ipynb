## Alpha Blending 

```python
import cv2
import numpy as np
```

```python
# Read the foreground image with alpha channel
foreGroundImage = cv2.imread("D:/Alphablending/foreGround.png", -1)
cv2.imshow('foreGround Image', foreGroundImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Read background image
background = cv2.imread("D:/Alphablending/backGround.jpg")
cv2.imshow('Background Image', background)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
foreGroundImage.astype
```

```python
# Split png foreground image
b,g,r,a = cv2.split(foreGroundImage)

# Save the foregroung RGB content into a single object
foreground = cv2.merge((b,g,r))
cv2.imshow('Foreground Image without alpha channel', foreground)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Save the alpha information into a single Mat
alpha = cv2.merge((a,a,a))
```

```python
# Convert uint8 to float
foreground = foreground.astype(float)
background = background.astype(float)
alpha = alpha.astype(float)/255
```

```python
# Perform alpha blending
foreground = cv2.multiply(alpha, foreground)
cv2.imshow('Blending of Foreground Image with alpha channel', foreground)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
background = cv2.multiply(1.0 - alpha, background)
outImage = cv2.add(foreground, background)
```

```python
cv2.imwrite("outImg.png", outImage)
cv2.imshow("outImg", outImage/255)
cv2.waitKey(0)
```

## Mathematical operations on images - Contrast Enhancement, Brightness Enhancement (Image Enhancement Techniques)

### Adjusting brightness and contrast

```python
#Import the necessary libraries 
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
  
# Load the image 
image = cv2.imread('D:/image_examples/milky-way-5295160_640.jpg') 
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image_rgb) 
  
```

```python
# Adjust the brightness and contrast 
# Adjusts the brightness by adding 12 to each pixel value 
brightness = 12 
# Adjusts the contrast by scaling the pixel values by 2.5 
contrast = 2.5  
image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness) 
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB) 
#Save the image 
cv2.imwrite('modified_image.jpg', image2) 
#Plot the contrast image 
plt.subplot(1, 2, 2) 
plt.title("Brightness & contrast") 
plt.imshow(image2_rgb) 
plt.show()
```

```python
#Import the necessary libraries 
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
  
# Load the image 
image = cv2.imread('D:/image_examples/milky-way-5295160_640.jpg') 
  
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 
  
# Adjust the brightness and contrast  
# g(i,j)=Œ±‚ãÖf(i,j)+Œ≤ 
# this is control Contrast by 1.5    (A value of 1.5 means the contrast is increased by 50%.)
alpha = 1.5  
# brightness control parameter ...control brightness by 50   (A value of 50 means the brightness is increased by 50 units)
beta = 50  
image2 = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)  #g(i,j)=Œ±‚ãÖf(i,j)+Œ≤  
 #It applies the formula g(i, j) = Œ± * f(i, j) + Œ≤ to each pixel in the image, where f(i, j) is the pixel value in the original image, 
#Œ± is the contrast factor, and Œ≤ is the brightness factor.  
#Save the image 
cv2.imwrite('Brightness & contrast.jpg', image2) 
#Plot the contrast image 
plt.subplot(1, 2, 2) 
plt.title("Brightness & contrast") 
plt.imshow(image2) 
plt.show() 
```

### Enhancing color in images

```python
#Import the necessary libraries 
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
  
# Load the image 
image = cv2.imread('D:/image_examples/milky-way-5295160_640.jpg') 
  
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 
  
# Convert the image from BGR to HSV color space 
image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV) 
  
# Adjust the hue, saturation, and value of the image 
# Adjusts the hue by multiplying it by 0.7 
image[:, :, 0] = image[:, :, 0] * 0.7
# Adjusts the saturation by multiplying it by 1.5 
image[:, :, 1] = image[:, :, 1] * 1.5
# Adjusts the value by multiplying it by 0.5 
image[:, :, 2] = image[:, :, 2] * 0.5
  
# Convert the image back to BGR color space 
image2 = cv2.cvtColor(image, cv2.COLOR_HSV2BGR) 
  
#Save the image 
cv2.imwrite('enhanced coloured.jpg', image2) 
  
#Plot the enhanced image 
plt.subplot(1, 2, 2) 
plt.title("enhanced coloured") 
plt.imshow(image2) 
plt.show()
```

```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
  
# Load the image 
image = cv2.imread('D:/image_examples/milky-way-5295160_640.jpg') 
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  
# Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image_rgb)
  
# Convert the image from RGB to HSV color space 
image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
  
# Adjust the hue, saturation, and value of the image 
image_hsv[:, :, 0] = image_hsv[:, :, 0] * 0.7
image_hsv[:, :, 1] = image_hsv[:, :, 1] * 1.5
image_hsv[:, :, 2] = image_hsv[:, :, 2] * 0.5
  
# Convert the image back to RGB color space 
image2_rgb = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)
  
# Save the image 
cv2.imwrite('enhanced_coloured.jpg', cv2.cvtColor(image2_rgb, cv2.COLOR_RGB2BGR))
  
# Plot the enhanced image 
plt.subplot(1, 2, 2) 
plt.title("enhanced coloured") 
plt.imshow(image2_rgb) 
plt.show()
```

### Equalizing histograms 

```python
#Import the necessary libraries 
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
  
# Load the image 
image = cv2.imread('D:/image_examples/milky-way-5295160_640.jpg') 
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image_rgb) 
  
# Convert the image to grayscale 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
  
# Equalize the histogram 
equalized_image = cv2.equalizeHist(gray_image) 
  
#Save the equalized image 
cv2.imwrite('equalized.jpg', equalized_image) 
  
#Plot the equalized image 
plt.subplot(1, 2, 2) 
plt.title("equalized") 
plt.imshow(equalized_image) 
plt.show()
```
